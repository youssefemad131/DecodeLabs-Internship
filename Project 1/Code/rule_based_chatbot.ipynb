## Step 2: The Anti-Pattern vs The Professional Approach

### ❌ The IF-ELIF Ladder — The Anti-Pattern

```python
# BAD — O(n) linear complexity, hard to maintain
if user_input == "hello":
    print("Hi there!")
elif user_input == "bye":
    print("Goodbye!")
elif user_input == "what is ai":
    print("AI is...")
# ... 50 more elif blocks = cascading failure risk
```

### ✅ The Dictionary + `.get()` — The Professional Approach

```python
# GOOD — O(1) constant time, easy to scale
responses = {"hello": "Hi there!", "bye": "Goodbye!"}
reply = responses.get(user_input, "I don't understand.")
```

**Why it matters:** If-elif scans every condition top-to-bottom — O(n). A dictionary is a hash map — it goes directly to the answer in O(1) constant time, regardless of how many rules you have.

## Step 4: Phase 1 — Input Sanitization

> Raw input can be `"HeLLo"`, `"  HELLO  "`, `"hello"` — all should map to the same intent.
> `.lower().strip()` is our sanitization pipeline.

````python
def sanitize(raw_input: str) -> str:
    """Phase 1: Normalize raw input — strip whitespace, lowercase."""
    return raw_input.lower().strip()

# ── Proof of concept ──
test_inputs = ["HeLLo", "  HELLO  ", "hello", "GOOD MORNING", "  What Is AI  "]

print("Input Sanitization Demo:")
print("-" * 40)
for raw in test_inputs:
    clean = sanitize(raw)
    print(f"  Raw: {repr(raw):<25} → Clean: {repr(clean)}")
````

## Step 5: Phase 2 — Intent Matching Engine

> Uses `dict.get()` for O(1) exact match first, then partial keyword matching for natural language flexibility, then a fallback for unknown inputs.

````python
def get_response(clean_input: str) -> str:
    """
    Phase 2: Intent Matching.
    1. Exact match   → O(1) dict lookup  (fastest)
    2. Partial match → keyword scan      (flexible)
    3. Fallback      → default response  (graceful)
    """

    # ── 1. Exact match ──
    if clean_input in KNOWLEDGE_BASE:
        return random.choice(KNOWLEDGE_BASE[clean_input])

    # ── 2. Partial / keyword match ──
    for key in KNOWLEDGE_BASE:
        if key in clean_input or clean_input in key:
            return random.choice(KNOWLEDGE_BASE[key])

    # ── 3. Fallback ──
    fallbacks = [
        "I don't understand that yet. Try rephrasing! 🤔",
        "That's outside my current rules. Try: 'what is AI' or 'hello'! 💡",
        "Unknown input. My guardrails say: I cannot process that. Try something else! 🛡️",
    ]
    return random.choice(fallbacks)


# ── Test the engine ──
test_queries = [
    "hello",
    "what is ai",
    "tell me a joke",
    "what is knn",
    "what is decodelabs",
    "random unknown input xyz",
]

print("Intent Matching Demo:")
print("=" * 60)
for query in test_queries:
    response = get_response(sanitize(query))
    print(f"\nYou    : {query}")
    print(f"DecoBot: {response}")
````

## Step 6: Phase 3 — The Heartbeat: The Infinite Loop

> *"The organism stays alive until the Kill Command."*
>
> `while True` → get input → sanitize → match intent → output response → repeat  
> `if exit → break` (Kill Command)

**Note:** Run this cell to start the chatbot. Type `exit` or `quit` to stop it.

````python
def run_chatbot():
    """The full IPO chatbot loop."""

    # ── Boot ──
    print("\n" + "=" * 62)
    print("  🤖  DecoBot — Rule-Based AI Chatbot")
    print("  DecodeLabs | Project 1 | Batch 2026")
    print("=" * 62)
    print("  Commands: 'help' for options | 'exit' to quit")
    print("=" * 62 + "\n")
    print("DecoBot: Hello, Engineer! I am DecoBot — your deterministic AI.")
    print("         Ask me about AI, Python, or just say hi! 🚀\n")

    # ── THE HEARTBEAT — Infinite Loop ──
    while True:

        # PHASE 1: Raw Input
        try:
            raw_input_text = input("You: ")
        except (EOFError, KeyboardInterrupt):
            print("\nDecoBot: Emergency shutdown. Goodbye! 👋")
            break

        # PHASE 1: Sanitize
        clean = sanitize(raw_input_text)

        # Guard: empty input
        if not clean:
            print("DecoBot: (No input detected. Please type something.)\n")
            continue

        # KILL COMMAND — Exit Strategy (clean break)
        if clean in ("exit", "quit", "q", "bye", "goodbye"):
            print("\nDecoBot: Goodbye, Engineer! Keep building and keep learning. 👋")
            print("         DecoBot signing off. 🤖\n")
            print("=" * 62)
            break

        # HELP command
        if clean == "help":
            print("\nDecoBot: Here are some things you can ask:\n")
            helps = [
                "  → 'what is AI'           — AI concepts",
                "  → 'what is KNN'          — ML algorithms",
                "  → 'what is Python'       — programming",
                "  → 'what is DecodeLabs'   — about the program",
                "  → 'what is project 2'    — next project",
                "  → 'what time is it'      — current time",
                "  → 'tell me a joke'       — fun!",
                "  → 'exit' / 'quit'        — shut down",
            ]
            for h in helps:
                print(h)
            print()
            continue

        # PHASE 2: Match intent
        response = get_response(clean)

        # PHASE 3: Output
        print(f"\nDecoBot: {response}\n")


# ── Launch ──
run_chatbot()
````

## Step 7: Algorithmic Efficiency Analysis

Let's benchmark the difference between the anti-pattern (if-elif) and the professional approach (dictionary).

````python
import time
import matplotlib.pyplot as plt

def benchmark_approaches(sizes):
    """Compare if-elif O(n) vs dictionary O(1) performance."""

    elif_times = []
    dict_times  = []

    for n in sizes:
        # Build n-rule structures
        keys = [f"intent_{i}" for i in range(n)]
        lookup_dict = {k: f"response_{i}" for i, k in enumerate(keys)}
        target = keys[-1]  # worst case: last key

        TRIALS = 10_000

        # ── if-elif simulation (linear scan) ──
        start = time.perf_counter()
        for _ in range(TRIALS):
            for k in keys:        # simulates chained elif
                if k == target:
                    _ = lookup_dict[k]
                    break
        elif_times.append((time.perf_counter() - start) * 1000)

        # ── dictionary O(1) ──
        start = time.perf_counter()
        for _ in range(TRIALS):
            _ = lookup_dict.get(target, "fallback")
        dict_times.append((time.perf_counter() - start) * 1000)

    return elif_times, dict_times


sizes = [10, 50, 100, 500, 1000, 5000, 10000]
elif_t, dict_t = benchmark_approaches(sizes)

plt.figure(figsize=(10, 5))
plt.plot(sizes, elif_t, 'o-', color='#FF5722', linewidth=2.5,
         markersize=7, label='if-elif ladder  O(n) — LINEAR GROWTH')
plt.plot(sizes, dict_t, 's-', color='#4CAF50', linewidth=2.5,
         markersize=7, label='Dictionary .get()  O(1) — CONSTANT TIME')
plt.fill_between(sizes, elif_t, dict_t, alpha=0.12, color='red')
plt.xlabel('Number of Rules (Scale)', fontsize=12)
plt.ylabel('Time for 10,000 lookups (ms)', fontsize=12)
plt.title('Algorithmic Efficiency: if-elif O(n) vs Dictionary O(1)', fontsize=13, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nAt 10,000 rules:")
print(f"  if-elif  : {elif_t[-1]:.2f} ms")
print(f"  dict.get : {dict_t[-1]:.2f} ms")
print(f"  Speedup  : {elif_t[-1]/dict_t[-1]:.0f}x faster with dictionary!")
````
